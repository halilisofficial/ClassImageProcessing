```python
#TODO: create a function for changing filter
from typing import Reversible
from cv2 import imread,IMREAD_GRAYSCALE
from numpy import asarray
"""
applying filter to a given image
"""


#default filter
filter = [
[-1, 0, 1],
[-1, 0, 1],
[-1, 0, 1]
]

def change_filter(filtername):
  global filter
  """
  https://www.codingame.com/playgrounds/2524/basic-image-manipulation/filtering
  Sharpening
  The details of an image can be emphasized by using a high-pass filter:
  In this kernel, the pixel is boosted when the neighbor pixels are different.
  filter:
  a b c    a is main ref 
  d e f
  g h i
  :filter
  """
  if filtername == "Sharpening":
    filter = [
    [0, -0.5, 0],
    [-0.5, 3, -0.5],
    [0, -0.5, 0]
    ]
  elif filtername == "Convolution":
    filter = [
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
    ]

change_filter("Convolution")
#get image as gray scale
img = imread("/content/img.png",IMREAD_GRAYSCALE)
#creating filtered_img array
filtered_img = []
for i in range(0,len(img)-3):
  temp = []
  for j in range(0,len(img[0])-3):
    temp.append(0)
  filtered_img.append(temp)


#convert all variables to array
img = asarray(img)
filter = asarray(filter)
filtered_img = asarray(filtered_img)

#apply filter to matrix
for i in range(0,len(img)-len(filter),1):
  for j in range(0,len(img[0])-len(filter[0]),1):
    # filtered_img array = a*f(0,0) + b*f(0,1) + c*f(0,2) + d*f(1,0) ... + i*f(2,2)
    filtered_img[i,j] = img[i,j]*filter[0,0] + img[i,j+1]*filter[0,1] + img[i,j+2]*filter[0,2] + img[i+1,j]*filter[1,0] + img[i+1,j+1]*filter[1,1] + img[i+1,j+2]*filter[1,2] + img[i+2,j]*filter[2,0] + img[i+2,j+1]*filter[2,1] + img[i+2,j+2]*filter[2,2]
print("#####################################")
print("normal image".title())
print(img)
print("#####################################")
print("filtered image".title())
print(filtered_img)
print("#####################################")

#apply pooling to filtered image
#https://paperswithcode.com/method/average-pooling
"""
pool:
a b
c d
:pool
avg_pooled_array.append(avg([a,b,c,d]))
"""
avg_pooled_array = []
max_pooled_array = []
for i in range(0,len(filtered_img)-2,2):
  temp_for_max_pool = []
  for j in range(0,len(filtered_img)-2,2):
    try:
        avg_pooled_array.append(
            (filtered_img[i,j] + filtered_img[i,j+1] + filtered_img[i+1,j] + filtered_img[i+1,j+1])/4
        )
        temp_for_max_pool.append(filtered_img[i,j])
        temp_for_max_pool.append(filtered_img[i,j+1])
        temp_for_max_pool.append(filtered_img[i+1,j])
        temp_for_max_pool.append(filtered_img[i+1,j+1])
        max_pooled_array.append(max(temp_for_max_pool))
    except:
      continue
print("average pooled array".title())
print(avg_pooled_array)
print("#####################################")
print("Max Pooled Array".title())
print(max_pooled_array)
print("#####################################")
```
